## Muestreo de presa (o depósito)

```python
import numpy as np

class MuestreoPresa:
    def __init__(self, k):
        self.k = k
        self.i = 0
        self.presa = np.zeros(self.k)

    def __call__(self, x):
        if self.i < self.k:
            self.presa[self.i] = x
        else:
            j = np.random.randint(0, self.i)
            if j < self.k:
                self.presa[j] = x    
        self.i += 1

    def obtener_presa(self):
        return self.presa
```

Probamos nuestra función con un arreglo de números enteros del 0 al 199.

```python
flujo = np.arange(0, 200)
n = len(flujo)
k = 10
mp = MuestreoPresa(k)
for i in range(n):
    mp(flujo[i])
    print('{0}: Nuevo dato = {1} Muestras = {2}'.format(i, flujo[i], np.sort(mp.obtener_presa())))
```
